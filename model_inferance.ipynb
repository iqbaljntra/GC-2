# IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import joblib
```

# LOAD DATA

```python
with open('model_log_reg.pkl', 'rb') as file_1:
  model_log_reg= joblib.load(file_1)

with open('model_scaler.pkl', 'rb') as file_2:
  model_scaler=joblib.load(file_2)

with open('model_encoder.pkl', 'rb') as file_3:
  model_encoder= joblib.load(file_3)

with open('list_num_cols.txt', 'rb') as file_4:
  num_cols= joblib.load(file_4)

with open('list_cat_cols.txt', 'rb') as file_5:
  cat_cols= joblib.load(file_5)
```

### MAKING RANDOM VALUE BASED ON FEATURES

```python
random = np.random.default_rng(seed=2895)
capital_gain = random.integers(0, 99999, size=30)
education_num = random.uniform(1, 16, size=30)
hours_per_week = random.integers(0, 99, size=30)
occupation = random.choice([ ' Craft-repair', ' Transport-moving', ' Sales', ' Adm-clerical',
                             ' Other-service', ' Farming-fishing', ' Prof-specialty',
                             ' Exec-managerial', ' Protective-serv', ' Handlers-cleaners',
                             ' Priv-house-serv', ' Machine-op-inspct', ' Tech-support'], size=30)
sex = random.choice([' Female',' Male'], size=30)
data_inf = pd.DataFrame({'capital_gain' : capital_gain, 'education_num' : education_num, 'hours_per_week' : hours_per_week, 'sex' :sex, 'occupation':occupation})
data_inf
```

# MODEL

```python
data_inf_num = data_inf[num_cols]
data_inf_cat = data_inf[cat_cols]
```

```python
data_inf_num_scaled = model_scaler.transform(data_inf_num)
data_inf_cat_encoded = model_encoder.transform(data_inf_cat)
```

```python
data_inf_final = np.concatenate([data_inf_num_scaled, data_inf_cat_encoded], axis=1)
```

### Making prediction INCOME BRACKET

```python
y_pred_inf = model_log_reg.predict(data_inf_final)
y_pred_inf
```
